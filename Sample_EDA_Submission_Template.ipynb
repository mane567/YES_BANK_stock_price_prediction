# **Project Summary -**

This project presents a comprehensive Exploratory Data Analysis (EDA) of the historical stock price data of YES Bank, one of India’s well-known private sector banks. The dataset consists of daily stock price records over multiple years, including the Open, High, Low, and Close prices for each trading day. These variables capture the intraday and end-of-day price movements of the stock and serve as the foundation for understanding its overall market behavior.

The primary objective of this project is to analyze stock price patterns, identify long-term trends, detect volatility phases, and examine the relationships among different price variables. By performing systematic data cleaning, transformation, and visualization, the project aims to uncover meaningful insights about price fluctuations, growth periods, market corrections, and high-risk intervals. Understanding these aspects is essential for investors, analysts, and financial researchers who rely on historical trends to make informed decisions.

The analysis begins with data cleaning and preprocessing. This includes checking for missing values, identifying and removing duplicate records, and converting the date column into a proper datetime format for time-series analysis. The dataset is then sorted chronologically to ensure accurate trend evaluation. These preprocessing steps are crucial for maintaining data integrity and ensuring reliable analytical outcomes.

Next, a statistical summary of the stock prices is conducted to understand central tendencies and dispersion. Measures such as mean, median, minimum, maximum, and standard deviation provide a quick overview of the stock’s performance range and variability. This helps in identifying unusual price spikes or extreme values that may correspond to major market events or company-specific developments.

Time-based trend analysis forms a core part of the EDA. Line plots are used to visualize the movement of closing prices over time, highlighting long-term growth, decline phases, and sudden crashes. Monthly averages and rolling mean calculations are applied to smooth short-term fluctuations and reveal underlying trends. These visualizations make it easier to interpret how the stock evolved across different economic or financial periods.

Volatility and return analysis further enhance the understanding of risk associated with the stock. Daily returns are calculated to measure percentage changes in closing prices, and volatility patterns are observed through spread analysis (difference between High and Low prices). Periods with sharp spikes in returns or large spreads indicate increased uncertainty and higher trading risk. Such analysis is particularly useful for assessing market instability and investor sentiment.

Correlation analysis is also performed to explore relationships between Open, High, Low, and Close prices. Heatmaps and pair plots are used to visualize correlations and identify linear dependencies among variables. Typically, strong positive correlations are observed among these price features, reflecting the interconnected nature of daily trading values. Understanding these relationships helps in selecting features for future predictive modeling.

Various visualization techniques—including line plots, box plots, histograms, heatmaps, and pair plots—are employed to support data storytelling. Box plots help detect outliers and extreme price values, while histograms reveal distribution patterns. Together, these visual tools transform raw numerical data into interpretable insights.

Overall, this EDA project provides a detailed understanding of YES Bank’s stock price behavior, highlighting key trends, volatility phases, and feature relationships. The insights derived from this analysis establish a strong foundation for developing predictive models such as regression-based approaches or advanced time-series forecasting techniques in future phases of the project.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# ================================
# YES BANK STOCK DATA WRANGLING
# ================================

# 1️⃣ Import Required Libraries
import pandas as pd
import numpy as np

# 2️⃣ Load Dataset
df = pd.read_csv("data_YesBank_StockPrices.csv")

print("Initial Data Preview:")
print(df.head())
print("\nInitial Shape:", df.shape)


# 3️⃣ Basic Information
print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())


# 4️⃣ Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# If missing values exist → Handle them
df.dropna(inplace=True)   # Option: remove missing rows
# df.fillna(method='ffill', inplace=True)  # Alternative: forward fill


# 5️⃣ Remove Duplicate Rows
print("\nDuplicate Rows:", df.duplicated().sum())
df.drop_duplicates(inplace=True)


# 6️⃣ Convert Date Column to Datetime
df['Date'] = pd.to_datetime(df['Date'], format='%b-%y')


# 7️⃣ Sort Data Chronologically (Important for Time Series)
df.sort_values(by='Date', inplace=True)
df.reset_index(drop=True, inplace=True)


# 8️⃣ Ensure Correct Data Types
df[['Open', 'High', 'Low', 'Close']] = df[['Open', 'High', 'Low', 'Close']].astype(float)


# 🔟 Final Clean Dataset
print("\nFinal Data Preview:")
print(df.head())

print("\nFinal Shape:", df.shape)

print("\nData Wrangling Completed Successfully ✅")
````

What manipulations have you done?

#### Chart - 10- Correlation Heatmap

````python
plt.figure(figsize=(10,6))

# Select only numerical price columns
price_cols = ['Open', 'High', 'Low', 'Close']

# Compute correlation matrix
corr_matrix = df[price_cols].corr()

# Plot heatmap
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            linewidths=0.5)

plt.title("Correlation Heatmap of Stock Price Variables")
plt.show()
````

##### 1. Why did you pick the specific chart?

The Correlation Heatmap was chosen to quickly visualize and understand the strength and direction of relationships between all numerical stock variables in a single view.

##### 2. What is/are the insight(s) found from the chart?

The heatmap shows a strong positive correlation among Open, High, Low, and Close prices, indicating they move closely together, while features like Daily Return have weaker correlation with absolute price levels.

#### Chart - 11- Pair Plot

````python
# Create features if not already created
df['Spread'] = df['High'] - df['Low']
df['Daily_Return'] = df['Close'].pct_change()

sns.pairplot(df[['Close','Spread','Daily_Return']].dropna())
plt.show()
````

##### 1. Why did you pick the specific chart?

The Pair Plot was chosen to visually analyze the relationships between Close price, volatility (Spread), and Daily Return in a single comprehensive view.

##### 2. What is/are the insight(s) found from the chart?

It shows that volatility increases during major price movements while Daily Return has weaker correlation with absolute price levels.
